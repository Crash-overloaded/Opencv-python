```
import cv2
from google.colab.patches import cv2_imshow
```

```
# Reading image
# image name , image property
image=cv2.imread('amazon_wall.jpg',1)
image1=cv2.imread('amazon_wall.jpg',0)
# 1 means image is in same color channel r,g,b or b,g,r
# 0 means no color channel i.e black & white or known as GreyImage
# -1 maintains image transparency
print(image.shape) # It prints hieght , width , color channels-->(R,G,B)
#cv2_imshow(image[300:600,:])
# Splitting image in B,G,R
x,y,z=cv2.split(image)
#### Mrging it by changing values of B,G,R 


#image=np.full((500,500),254)
cv2_imshow(x)
cv2_imshow(y)
cv2_imshow(z)


# cv2.waitKey(0) -->  holds the image for infinite time till we close it...
# Press "q"  for quitting image window
```

```
# drawing line on image
# syntax --> image name ,starting point , ending point , width
cv2.line(image,(500,100),(100,500),(0,0,0),3)
cv2.line(image,(100,100),(500,500),(255,255,255),3)
cv2.line(image,(300,100),(300,500),(0,0,255),3)
cv2.line(image,(100,300),(500,300),(255,0,0),3)
cv2.rectangle(image,(100,100),(500,500),(0,255,0),3)
cv2_imshow(image)
```

```
# Saving image
cv2.imwrite('ama.jpg',image1)
cv2.imread('ama.jpg')
```
